```python
import pandas as pd
```

```python
# Importing data from excel
rawDf = pd.read_csv('data_cleaning_challenge.csv')
rawDf.head()
```

```python
# Dropping null first name values
filledDF = rawDf.drop(columns=['Unnamed: 9', 'Unnamed: 10'])
filledDF = filledDF[filledDF['Row Type'].notna()]

# Iterating colums by first name
column_values = []
counter = 0

for i in filledDF['Row Type']:
    if "first name" in i:
        counter += 1
    column_values.append(counter)

filledDF['Iteration'] = column_values

# Removing unwanted rows
filledDF = filledDF[filledDF['Row Type'] != 'Row Type']

# Selecting only rows related to first name, last name and date
nameDF = filledDF.copy()
nameDF.drop(columns=['Speed1','Speed2','Electricity','Effort','Weight','Torque'], inplace=True)
nameDF.rename(columns={'Row Type': 'First Name', 'Iter Number': 'Last Name', 'Power1': 'Date'}, inplace=True)
nameDF = nameDF[nameDF['First Name'].str.contains('first name')]

nameDF['First Name'] = nameDF['First Name'].apply(lambda name: name.split(':')[1])
nameDF['Last Name'] = nameDF['Last Name'].apply(lambda name: name.split(':')[1])
nameDF['Date'] = nameDF['Date'].apply(lambda name: name.split(':')[1])

# Filling null values in stats
statsDF = filledDF.copy()
statsDF.fillna(value=0, inplace=True)

# Merging both final DataFrames 
finalDF = pd.merge(left=nameDF, right=statsDF, how='inner', on='Iteration')
finalDF = finalDF[(finalDF['Speed1'] != 0) & (finalDF['Speed2'] != 0) & (finalDF['Electricity'] != 0) & (finalDF['Effort'] != 0) & (finalDF['Weight'] != 0) & (finalDF['Torque'] != 0)]

# Exporting cleaned data to Excel
finalDF.to_csv('dataCleaned.csv')
finalDF

```
